```python
import random
import numpy as np

print("Please select a number 1-3 representing the difficulty")
print("1 represents easy difficulty. 2 represents medium difficulty. 3 represents hard difficulty")
game_style = int(input("Please enter a number between 1 and 3: "))
print(" ")

if game_style == 1:
    z = 25
    n = 5
    mines = 5
if game_style == 2:
    z = 36
    n = 6
    mines = 7
if game_style == 3:
    z = 49 
    n = 7
    mines = 10

board = np.arange(z).reshape((n,n))
board *= 0

board_real = np.arange(z).reshape((n,n))
board_real = (board_real*0) + 9

num = 0
while num < mines:
    row = random.randint(0, n - 1)
    column = random.randint(0, n - 1)
    if board[row][column] == 0:
        board[row][column] = 1
        num += 1

def nearby_mines(row, column):
    total_mines = 0
    i = row - 1
    while i <= row + 1:
        if i >= 0 and i < n:
            j = column - 1
            while j <= column + 1:
                if j >= 0 and j < n:
                    total_mines = total_mines + board[i][j]
                j += 1
        i += 1
    board_real[row][column] = total_mines
    return total_mines
    
def display():
    for row in range(0, n):
        for column in range(0, n):
            print(board[row][column], end = " ")
        print("")
        
def display_real():
    if n == 5:
        print("-"*21)
    if n == 6:
        print("-"*25)
    if n == 7:
        print("-"*29)
    for row in range(0, n):
        print("| ", end = "")
        for column in range(0, n):
            if board_real[row][column] == 9:
                print(" ", end = " | ")
            else:
                print(nearby_mines(row, column), end = " | ")
        print("")
        if n == 5:
            print("-"*21)
        if n == 6:
            print("-"*25)
        if n == 7:
            print("-"*29)   
                    
player = 0         
while player <= (z - mines):
    guess_row = int(input("Enter a row: "))
    guess_column = int(input("Enter a column: "))
    if board[guess_row - 1][guess_column - 1] == 1:
        print("You Lose")
        break
    else:
        board_real[guess_row - 1][guess_column - 1] = nearby_mines(guess_row - 1, guess_column - 1)
        player += 1
        display_real() 
    if player == (z - mines):
        print("You Win!")
        break
        

display()
```
